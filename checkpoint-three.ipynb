## Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python
df.describe()
#figuring out if outliers by using min, max, and mean
```

```python
numeric_cols = ["$Worldwide", "$Domestic", "$Foreign", 
                "Domestic %", "Foreign %", "Vote_Count"]

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"Outliers for {col}:")
    print(outliers)
 
#trying to get a better understanding of outliers in each column
```

```python

plt.boxplot(df["$Worldwide"])
plt.title("Worldwide Revenue Outliers")
plt.show()
```
